```python
from datetime import date
import datetime
from eth_tools.room_allocation.scraper import download_room_allocation


for i, room in enumerate(global_room_info["rooms"]):
    print(f"Downloading room {i+1}/{len(global_room_info['rooms'])}")
    download_room_allocation(
        room=f"{room['building']} {room['floor']} {room['room']}", 
        from_date=date.today().isoformat(), 
        to_date=(date.today() + datetime.timedelta(days=7)).isoformat()
    )
```
